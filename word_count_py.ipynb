```python
#!/usr/bin/env python3

import sys

def main():
    # Example task: Count the frequency of each word from stdin, streaming input line by line.
    # This is fast, lightweight, and uses only the standard library.

    from collections import Counter

    word_counts = Counter()
    for line in sys.stdin:
        words = line.strip().split()
        word_counts.update(words)

    for word, count in word_counts.most_common():
        print(f"{word}\t{count}")

if __name__ == "__main__":
    main()
```
